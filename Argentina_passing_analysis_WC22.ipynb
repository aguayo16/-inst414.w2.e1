```python
import pandas as pd
import networkx as nx
from statsbombpy import sb
import numpy as np
```

```python
sb.competitions()
```

```python
matches = sb.matches(competition_id=43, season_id=106)
matches = matches[(matches['home_team']=='Argentina') | (matches['away_team']=='Argentina')]
matches
```

```python
match1 = sb.events(match_id=3869685)
match2 = sb.events(match_id=3869151)
match3 = sb.events(match_id=3869321)
match4 = sb.events(match_id=3857264)
match5 = sb.events(match_id=3857289)
match6 = sb.events(match_id=3869519)
match7 = sb.events(match_id=3857300)

matches_list = [match1,match2,match3,match4,match5,match6,match7]
```

```python
total_matches = pd.DataFrame()
for match in matches_list:
    total_matches = total_matches.append(match)
total_matches
```

```python
total_matches.columns
```

```python
df = total_matches[['team','player_id','player','pass_outcome','pass_recipient']]
df = df[df['team']=='Argentina'].reset_index()
df = df[df['pass_recipient'].notnull()]
#df = df[df['pass_outcome'].notnull()==False]
df
```

```python
data = {
  "passer": df['player'],
  "receiver": df['pass_recipient']
}

pass_counts = pd.DataFrame(data).reset_index(drop=True)
pass_counts
```

```python
pass_counts = pass_counts.groupby(['passer','receiver']).size().reset_index(name='count')
pass_counts
```

```python
G = nx.from_pandas_edgelist(pass_counts,
                           source = 'passer',
                           target = 'receiver',
                           edge_attr = 'count',
                           create_using = nx.DiGraph())
nx.draw(G, with_labels=True)
```

```python
nx.write_graphml(G, "passes.graphml", prettyprint=False)
```

```python
centrality_degree = nx.degree_centrality(G)
centrality_degree 
```

```python
# sort node-centrality dictionary by metric, and reverse to get top elements first
for u in sorted(centrality_degree, key=centrality_degree.get, reverse=True)[:10]:
    print(u, centrality_degree[u])
```
