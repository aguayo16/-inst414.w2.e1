```python
res2 = requests.get('https://api.football-data-api.com/league-players?key=example&season_id=6135')
players=json.loads(res2.text)
players
```

```python
count = 0
name = []
minutes_played = []
appearances = []


d=[]
goals = []
for player in players['data']:
    name.append(player['full_name'])
    d.append(
        {
         'Minutes': player['minutes_played_overall'],
         'Appearances': player['appearances_overall'],
         'Goals': player['goals_overall'],
         'Assists': player['assists_overall'],
         'Avg Playing time': player['min_per_match'],
         'Yellow Cards': player['yellow_cards_overall'],
         'Red Cards': player['red_cards_overall'],
                   
        }
    )
    
df = pd.DataFrame(d, index=name)

df
```

```python
ndf = df.subtract(df.min()).divide(df.max()-df.min())
ndf
```

```python
query = ['Paul Pogba','Leandro Trossard','Gabriel Teodoro Martinelli Silva']
query_stats = ndf.loc[query[0]]

for i in query:
    query_stats = ndf.loc[i]
    distances = scipy.spatial.distance.cdist(ndf, [query_stats], metric='euclidean')[:,0]
    distances_from_query = list(zip(ndf.index, distances))
    distances_from_query

    for similar_player, similar_feature_score in sorted(distances_from_query, key=lambda x: x[1], reverse=False)[:10]:
        print(similar_player, similar_feature_score, df.loc[similar_player].sum())
    print('\n')
```
